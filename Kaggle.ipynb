```python
def diagnose_mv(df, mv_column):
    cols = df.columns
    flags = df[mv_column].isna().fillna(False)
    fig, ax = plt.subplots(len(cols), 3, 
                           figsize=(len(cols)+3, len(cols)+3), 
                           constrained_layout=True)
    plt.rcParams['axes.grid'] = True
    for i, col in enumerate(cols):
        n1, bins, _ = ax[i, 0].hist(df[col])
        ax[i, 0].set_title(f'{col} with MV')
        
        n2, _, _ = ax[i, 1].hist(df[col][~flags], bins=bins)
        ax[i, 1].set_title(f'{col} without MV')
        
        ax[i, 2].bar(bins[:-1], np.abs(n2-n1))
        ax[i, 2].set_title(f'Difference')
        
```

```python
# GSS code system of England and Wales
# Distribution of casualties in England and Wales:

# Custom function to map values to codes
def map_lsoa_to_code(value):
    if value.startswith("E"):
        return 1
    elif value.startswith("W"):
        return 2
    else:
        return 0  # or handle other cases if needed

# Apply the custom function to create a new 'GSS_code' column
df_pure['GSS_code'] = df_pure['lsoa_of_casualty'].apply(map_lsoa_to_code)

# Count the occurrences of each code
code_counts = df_pure['GSS_code'].value_counts()

# Plot the results
plt.figure(figsize=(4, 3))
sns.countplot(x='GSS_code', hue='GSS_code', data=df_pure)
plt.title('Distribution of casualties in England and Wales')
plt.xlabel('GSS_Code')
plt.ylabel('Count')
plt.legend(labels=['England', 'Wales'])
plt.show()

print(code_counts)
```

# Heat Map

```python
# Compute the correlation matrix
corr = df_pure.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
plt.figure(figsize=(15, 12))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True)

plt.title('Correlation Matrix Heatmap')
plt.show()
```
