# Missing Data

```python
df_pure = df.copy()
```

```python
df_pure = df_pure.replace(-1, pd.NA)  # Replace -1 with missing values (NA) 
df_pure = df_pure.where(df_pure != "-1", pd.NA)  # Replace "-1" with missing values (NA)
df_pure.isna().sum()
```

```python
#df_pure.dropna(inplace=True)  # Remove rows with missing values (NA)
```

# Distributions:

```python
df_pure = df_pure.sort_values(by='accident_index')
```

```python
print(df_pure['casualty_reference'].value_counts().head(30))

# Plot the distribution of casualties involved in an accident
plt.figure(figsize=(4, 3))
sns.countplot(x='casualty_reference', data=df_pure)
plt.title('Distribution of Accident Casualties')
plt.xlabel('Casualty Severity')
plt.ylabel('Count')
plt.show()
```

```python
# Plot the distribution of casualty classes
plt.figure(figsize=(4, 3))
sns.countplot(x='casualty_class', hue='casualty_class', data=df_pure)
plt.title('Distribution of casualty classes')
plt.xlabel('Casualty Class')
plt.ylabel('Count')
plt.legend(labels=['Driver or rider', 'Passenger', 'Pedestrian'])
plt.show()
```

```python
print(df_pure['sex_of_casualty'].value_counts())

# Plot the distribution of casualties sex
plt.figure(figsize=(4, 3))
sns.countplot(x='sex_of_casualty', hue='sex_of_casualty', data=df_pure)
plt.title('Distribution of casualties sex')
plt.xlabel('Sex of Casualty')
plt.ylabel('Count')
plt.legend(labels=['Male', 'Female', 'Unkown'])
plt.show()
```

```python
# Plot the distribution of casualties' age band
plt.figure(figsize=(8, 6))
sns.countplot(x='age_band_of_casualty', hue='age_band_of_casualty', data=df_pure)
plt.title("Distribution of casualties' age band")
plt.xlabel('Age band of Casualty')
plt.ylabel('Count')
plt.legend(labels=['0-5', '6-10', '11-15', '16-20', '21-25', '26-35', '36-45', '46-55', '56-65', '66-75', 'over 75'])
plt.show()
```

```python
# Plot the distribution of accident severities
plt.figure(figsize=(4, 3))
sns.countplot(x='casualty_severity', hue='casualty_severity', data=df_pure)
plt.title('Distribution of Accident Severities')
plt.xlabel('Casualty Severity')
plt.ylabel('Count')
plt.legend(labels=['Fatal', 'Serious', 'Slight'])
plt.show()
```

```python
# Plot the distribution of casualty types
plt.figure(figsize=(6, 4))
sns.countplot(x='casualty_type', data=df_pure)
plt.title('Distribution of Casualty Types')
plt.xlabel('Casualty Type')
plt.ylabel('Count')
plt.show()
```

```python
# Plot the distribution of casualties based on their casualty IMD decile.
plt.figure(figsize=(8, 6))
sns.countplot(x='casualty_imd_decile', hue='casualty_imd_decile', data=df_pure)
plt.title('Distribution of casualty IMD decile')
plt.xlabel('Casualty IMD decile')
plt.ylabel('Count')
plt.legend(labels=['Most deprived 10%', '10-20%', '20-30%', '30-40%', '40-50%',
                   'Less deprived 40-50%', '30-40%', '20-30%', '10-20%', 'Least deprived 10%'])
plt.show()
```

```python
# Plot the distribution of casualties based on the type of home area
plt.figure(figsize=(4, 3))
sns.countplot(x='casualty_home_area_type', hue='casualty_home_area_type', data=df_pure)
plt.title('Distribution of Casualty Home Area Type')
plt.xlabel('Casualty Home Area Type')
plt.ylabel('Count')
plt.legend(labels=['Urban area', 'Small town', 'Rural'])
plt.show()
```

```python
# GSS code system of England and Wales
# Distribution of casualties in England and Wales:

# Custom function to map values to codes
def map_lsoa_to_code(value):
    if value.startswith("E"):
        return 1
    elif value.startswith("W"):
        return 2
    else:
        return 0  # or handle other cases if needed

# Apply the custom function to create a new 'GSS_code' column
df_pure['GSS_code'] = df_pure['lsoa_of_casualty'].apply(map_lsoa_to_code)

# Count the occurrences of each code
code_counts = df_pure['GSS_code'].value_counts()

# Plot the results
plt.figure(figsize=(4, 3))
sns.countplot(x='GSS_code', hue='GSS_code', data=df_pure)
plt.title('Distribution of casualties in England and Wales')
plt.xlabel('GSS_Code')
plt.ylabel('Count')
plt.legend(labels=['England', 'Wales'])
plt.show()

print(code_counts)
```

```python
# Plot the relationship between casualty severity and gender
plt.figure(figsize=(4, 3))
sns.countplot(x='casualty_severity', hue='sex_of_casualty', data=df_pure)
plt.title('Casualty Severity by Gender')
plt.xlabel('Casualty Severity')
plt.ylabel('Count')
plt.legend(labels=['Male', 'Female'])
plt.show()
```

```python
# Plot the relationship between casualty severity and casualty class
plt.figure(figsize=(4, 3))
sns.countplot(x='casualty_severity', hue='casualty_class', data=df_pure)
plt.title('Casualty Severity by Casualty Class')
plt.xlabel('Casualty Severity')
plt.ylabel('Count')
plt.legend(labels=['Driver', 'Passenger', 'Pedestrian'])
plt.show()
```

```python
# Plot the relationship between casualty severity and casualty's home area type
plt.figure(figsize=(4, 3))
sns.countplot(x='casualty_severity', hue='casualty_home_area_type', data=df_pure)
plt.title('Casualty Severity by Home Area Type')
plt.xlabel('Casualty Severity')
plt.ylabel('Count')
plt.legend(labels=['Urban', 'Small town', 'Rural'])
plt.show()
```

# Heat Map

```python
# Compute the correlation matrix
corr = df_pure.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
plt.figure(figsize=(15, 12))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True)

plt.title('Correlation Matrix Heatmap')
plt.show()
```
